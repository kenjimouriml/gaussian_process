```python
import numpy as np
from scipy.stats import multivariate_normal
import math as m

from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(start=0.0, stop=10, num=40)
```

```python
def K(x):
    theta1 = 1.0
    theta2 = 1.0
    
    kernel_matrix = np.zeros((x.shape[0], x.shape[0]))
    for i in range(x.shape[0]):
        for j in range(x.shape[0]):
            kernel_matrix[i, j] = theta1 * m.exp(-np.dot(x[i] - x[j], x[i] - x[j]) / theta2)
    
    return kernel_matrix
```

```python
mean = np.zeros(x.shape[0])
cov = K(x)
```

```python
f = multivariate_normal.rvs( mean=mean, cov=cov, size=1)
```

```python
plt.plot(x, f)
```

```python
N = 500 * 2
x = np.random.rand(N).reshape(500 , 2)
x = 0.1 * np.arange(N).reshape(500 , 2)
```

```python
mean = np.zeros(x.shape[0])
cov = K(x)
```

```python
f = multivariate_normal.rvs( mean=mean, cov=cov, size=1)
```

```python
f = f.reshape(-1, 1)
```

```python
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(x[:, 0], x[:, 1], f)
```
